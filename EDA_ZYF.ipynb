```python
import pandas as pd
import numpy as np
from ipyleaflet import Map, Marker, MarkerCluster, Icon
```

```python
df_kiosk = pd.read_csv('Austin_B-Cycle_Kiosk_Locations.csv')
df_kiosk_cp = df_kiosk.copy()
df_kiosk_cp.head()
```

```python
# filter out non-active kiosk
df_kiosk_cp = df_kiosk_cp[df_kiosk_cp["Kiosk Status"] == "active"]
```

```python
df_kiosk_cp.info()
```

```python
df_fixit = pd.read_csv('Bike_Fix_It.csv')
df_fixit_cp = df_fixit.copy()
df_fixit_cp.head()
```

```python
# extract latitude and longitude from Location 1 column
df_fixit_cp['latitude'] = df_fixit_cp['Location 1'].apply(lambda x: x.split('\n')[2].strip(' ()').split(',')[0]).astype(float)
df_fixit_cp['longitude'] = df_fixit_cp['Location 1'].apply(lambda x: x.split('\n')[2].strip(' ()').split(',')[1]).astype(float)
```

```python
df_fixit_cp.head()
```

```python
center = (30.2672, -97.7431)

m = Map(center=center, zoom=15)

markera = []
for idx,row in df_fixit_cp[['latitude','longitude']].iterrows():
    mk = Marker(location=(row['latitude'], row['longitude']))
    markera.append(mk)

marker_cluster = MarkerCluster(
    markers=tuple(markera)
)
m.add_layer(marker_cluster);

m
```

```python
icon = Icon(icon_url='https://leafletjs.com/examples/custom-icons/leaf-green.png', icon_size=[38, 95], icon_anchor=[22,94])
markera2 = []
for idx,row in df_kiosk_cp[['Latitude','Longitude']].iterrows():
    mk2 = Marker(location=(row['Latitude'], row['Longitude']),icon=icon)
    markera2.append(mk2)

marker_cluster2 = MarkerCluster(
    markers=tuple(markera2)
)

m.add_layer(marker_cluster2);
display(m)
```
