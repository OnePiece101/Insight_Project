## YSL
* https://www.sephora.com/product/yves-saint-laurent-tatouage-couture-velvet-cream-liquid-lipstick-P457250

```python
pdlist = soup.find_all('div',class_='css-1ax77m2')
```

```python
lipsticks = []
opener=urllib.request.build_opener()
opener.addheaders=[('User-Agent','Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1941.0 Safari/537.36')]
urllib.request.install_opener(opener)
for i in range(len(pdlist)):
    brand = soup.find('h1',class_='css-140z8k4').find_all('span')[0].string
    looks = soup.find('h1',class_='css-140z8k4').find_all('span')[1].string
    product_id = pdlist[i].img['src'].split('/')[3].strip('s+w.jpg')
    color_img_url = 'https://www.sephora.com/productimages/sku/s'+product_id+'+sw.jpg'
    color_code = ' '.join(pdlist[i].button['aria-label'].split()[:3])
    color = ' '.join([e for e in pdlist[i].button['aria-label'].split()[3:] if e not in ['-','Selected']])
    lipsticks.append({'product_id':product_id,
                        'brand': brand,
                        'looks': looks,                         
                         'color_img_url': color_img_url,
                         'color_code': color_code,
                         'color': color})
    filename = '/home/yafen/insight_project/Week1Demo/img/'+product_id+'_swatch.jpg'
    urllib.request.urlretrieve(color_img_url,filename)
df_ysl = pd.DataFrame(lipsticks)
```
