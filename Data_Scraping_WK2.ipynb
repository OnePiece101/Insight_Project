```python
from bs4 import BeautifulSoup
import requests
```

```python
web = requests.get('https://www.sephora.com/product/lip-maestro-liquid-lipstick-P393411?skuId=2242709')
soup = BeautifulSoup(web.content,'lxml')
```

```python
print(soup.prettify())
```

## Tom Ford

```python
# picture of color on lips, url: https://www.sephora.com+output
soup.find_all('div',attrs={'class':'css-69norl'})[1].img['srcset']
```

```python
# img of lipstick, url: https://www.sephora.com+output
soup.find_all('div',attrs={'class':'css-69norl'})[0].img['src']
```

```python
# brand
soup.find('h1',class_='css-140z8k4').find_all('span')[0].string
```

```python
# looks
soup.find('h1',class_='css-140z8k4').find_all('span')[1].string
```

```python
# price
soup.find('div',class_='css-slwsq8').span.string
```

```python
# icon img of color palette, number between 's' and '+sw' is the same as number after 'skuId=' in product url
# and every page have all product id number of the brand
soup.find_all('div',class_='css-1ax77m2')[0].img['src']
```

```python
# color of the lipstick
soup.find('span',class_='css-ng5oyv').contents[2].split('-')[0]
```
