```python
import cv2
import os
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
path = '/home/yafen/insight_project/Week1Demo/img'
colors = []
for i in os.listdir(path):
    product_id = i.split('_')[0].strip()
    img_file = os.path.join(path,i)
    image = cv2.imread(img_file)
    rgb_mean = cv2.mean(image)[:3]
    colors.append({'product_id': product_id,
                        'rgb_mean': rgb_mean})
df_color = pd.DataFrame(colors)
```

```python
df_ysl = pd.read_csv('ysl.csv')
# df_ysl['product_id'] = df_ysl.product_id.astype('object')
df_color.product_id = df_color.product_id.astype(int)
```

```python
df_ysl = df_ysl.merge(df_color,on='product_id')
df_ysl
```

```python
df_ysl.to_csv('ysl.csv',index=False)
```

```python
Z = np.array([e for e in df_ysl.rgb_mean.values],dtype='float32')
kmeans = KMeans(n_clusters=5,init='random',random_state=0).fit(Z)
```

```python
df_ysl['color_label'] = kmeans.labels_
```

```python
label = kmeans.predict(np.reshape(np.array([28.01298701298701, 51.099567099567096, 222.42943722943724],dtype='float32'),(1,-1)))
```
